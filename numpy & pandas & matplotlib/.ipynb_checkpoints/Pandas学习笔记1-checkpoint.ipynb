```python
import pandas as pd
import numpy as np

# dataFrame
data = {'name': ['Joe', 'Mike', 'Jack', 'Rose', 'David', 'Marry', 'Wansi', 'Sidy', 'Jason', 'Even'],
'age': [25, 32, 18, np.nan, 15, 20, 41, np.nan, 37, 32],
'gender': [1, 0, 1, 1, 0, 1, 0, 0, 1, 0],
'isMarried': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(data, index=labels)

display(df)
```

```python
display(df.index)

display(df.values)

display(df.columns)

display(df.shape)

display(df.ndim)

display(df.dtypes)
```

```python
#  read_csv
import numpy as np
import pandas as pd
data = pd.read_csv(r'./directory.csv',header=0) #header=0header=None
# 
data.head(10)

#                        
```

```python
#
data_test = data[['Store Number','City']] # data[]
print('-'*50)
print(data_test.loc[0]) #0
print('-'*50)
print(data_test.loc[0:10]) #rangerange
print('-'*50)
print(data_test.loc[0:3,['City']]) #0-3City
print('-'*50)
print(data_test[0:3][['City']]) # 
print('-'*50)
print(data_test[['City']][0:3]) # 
print('-'*50)

# 
# []&|,~
print(data_test[   (data_test['City'] == "ShangHai") & (data_test['Store Number'] == "49020-266506") ])
print('-'*20)
print(data_test[   (data_test['City'] == "ShangHai") | (data_test['Store Number'] == "49020-266506") ])
print('-'*20)
print(data_test[   ~(data_test['City'] == "ShangHai") ])
print('-'*50)

# 
print(data_test[(data_test['City'] == "ShangHai")][['City']])
# data_test
```
