# Pandas数据类型
## Pandas 的数据结构：
* Pandas 主要有 Series（一维数组）
* DataFrame（二维数组）
* Panel（三维数组）
* Panel4D（四维数组
* PanelND（更多维数组）等数据结构。   


其中 Series 和 DataFrame 应用的最为广泛。  

* Series：Series是一维带标签的数组，它可以包含任何数据类型。包括整数，字符串，浮点数，Python 对象等。Series 可以通过标签来定位。 Seires中的数据，只能是一维的

* DataFrame：DataFrame是二维的带标签的数据结构。我们可以通过标签来定位数据。这是 NumPy 所没有的。 DataFrame中的数据一定得是二维的。 当然，其实一维也是一个特殊的二维

Pandas是一个开源的Python数据分析库。Pandas把结构化数据分为了三类：
* Series，1维序列，可视作为没有column名的、只有一个column的DataFrame；
* DataFrame，同Spark SQL中的DataFrame一样，其概念来自于R语言，为多column并schema化的2维结构化数据，可视作为Series的容器（container）；
* Panel，为3维的结构化数据，可视作为DataFrame的容器；

```python
# 当数据源为第三方数据时，读取数据可以自动变成带表头和索引的二维表格形式，读取数据 read_csv
import numpy as np
import pandas as pd
data = pd.read_csv(r'./directory.csv',header=0) #默认header=0，是包含表头的意思，header=None时，是不包含表头的意思。
# 默认显示前五条
data.head(10)

# 品牌  店里电话  店名  企业类别   街道地址  城市  省份  国家  邮编  地区  经度  纬度
```

```python
#取行取列
data_test = data[['Store Number','City']] # 如果拿data里的某几列，需要使用[]引起来，使其为列表形式。
print('-'*50)
print(data_test.loc[0]) #打印索引为0的这一行
print('-'*50)
print(data_test.loc[0:10]) #注意和range的规则不一样，这里是左闭右闭区间，range（）是左开右闭区间
print('-'*50)
print(data_test.loc[0:3,['City']]) #取0-3行，City列
print('-'*50)
print(data_test[0:3][['City']]) # 效果同上，可以在第二个列表中添加多列
print('-'*50)
print(data_test[['City']][0:3]) # 效果同上，逗号“，”代表取列，冒号“：”代表取行，是不分顺序的
print('-'*50)

# 条件过滤
# 如果要在[]里放条件，进行条件运算，那么要带着表的列进行运算，与的符号&，或的符号|,非得符号~
print(data_test[   (data_test['City'] == "ShangHai") & (data_test['Store Number'] == "49020-266506") ])
print('-'*20)
print(data_test[   (data_test['City'] == "ShangHai") | (data_test['Store Number'] == "49020-266506") ])
print('-'*20)
print(data_test[   ~(data_test['City'] == "ShangHai") ])
print('-'*50)

# 取行取列可以和条件过滤同时使用
print(data_test[(data_test['City'] == "ShangHai")][['City']])
# data_test
```
