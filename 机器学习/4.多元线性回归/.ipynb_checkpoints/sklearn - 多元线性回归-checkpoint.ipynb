```python
import numpy as np
from numpy import genfromtxt
from sklearn import linear_model
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 读数据
data = genfromtxt("data.csv",delimiter=',')
data
```

```python
x_data = data[:,:-1] # 取除了倒数第一列外的所有列
y_data = data[:,-1] # 取倒数第一列
x_data
y_data
```

```python
# 建模
model = linear_model.LinearRegression()
model.fit(x_data,y_data)
```

```python
print("coefficients:",model.coef_)
print("coefficients:",model.intercept_)
```

```python
# 测试
x_test = [[104,3]]
pridict = model.predict(x_test)
print("pridict:",pridict)
```

```python
# 画图
ax = plt.figure().add_subplot(111,projection = '3d') # add_subplot（‘3d’）画3D图的函数
ax.scatter(x_data[:,0], x_data[:,1], y_data, c = 'r', marker = 'o', s = 10)
x0 = x_data[:,0]
x1 = x_data[:,1]
# 生成矩阵网络
x0,x1 = np.meshgrid(x0,x1)
z = model.intercept_ + x0 * model.coef_[0] + x1 * model.coef_[1]
# 画3D图
ax.plot_surface(x0,x1,z)
ax.set_xlabel('Miles')
ax.set_ylabel('Num of Deliveries')
ax.set_zlabel('Time')

plt.show()
```
