```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
data = np.genfromtxt("data.csv",delimiter=',')
x_data = data[1:,1]
y_data = data[1:,2]
plt.scatter(x_data,y_data)
plt.show()
data
```

```python
x_data = data[1:,1,np.newaxis]
y_data = data[1:,2,np.newaxis]
# 创建并拟合模型
model = LinearRegression()
model.fit(x_data,y_data)
```

```python
# 画图（错误尝试，测试）
plt.plot(x_data,y_data,'b.')
plt.plot(x_data,model.predict(x_data),'r')
plt.show()
```

```python
# 定义多项式回归，degree的值可以调节多项式的特征
poly_reg = PolynomialFeatures(degree=5)
# 特征处理
x_poly = poly_reg.fit_transform(x_data)
# 定义回归模型
lin_reg = LinearRegression()
# 训练模型
lin_reg.fit(x_poly,y_data)
```

```python
x_poly
```

```python
# 画图
plt.plot(x_data,y_data,'b.')
plt.plot(x_data,lin_reg.predict(x_poly),'r') # 10个点的话，曲线是不平滑的
plt.show()
```

```python
x_test = np.linspace(1,10,100) #从1到10均匀生成100个点
x_test = x_test[:,np.newaxis]
```

```python
# 平滑曲线画图
plt.plot(x_data,y_data,'b.')
plt.plot(x_test,lin_reg.predict(poly_reg.fit_transform(x_test)),'r') # 10个点的话，曲线是不平滑的
plt.show()
```
